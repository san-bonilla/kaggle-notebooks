```python
import pandas as pd

def find_employees(employee: pd.DataFrame) -> pd.DataFrame:
    # Subset employee by
    # Rows that have an employee salary higher than
    # ... Use map function to map the managerId to their respective salary
    # Return the rows where the salary of the employee is higher than the salary of their manager
    result_df = employee[employee['salary'] > employee['managerId'].map(employee.set_index('id')['salary'])]
    # Return the 'name' column as 'Employee'
    result_df = result_df[['name']].rename(columns={'name': 'Employee'})
    print(employee)
    return result_df

find_employees(df)
```
